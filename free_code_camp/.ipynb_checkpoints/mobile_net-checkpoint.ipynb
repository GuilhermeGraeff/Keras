```python
def prepare_image(file):
    img_path = '/home/ggraeff/Documents/Keras/free_code_camp/data/MoboleNet-samples/'
    img = image.load_img(img_path + file, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return tf.keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

```python
train_path = '/home/ggraeff/Documents/Keras/free_code_camp/data/sign-language/train'
valid_path = '/home/ggraeff/Documents/Keras/free_code_camp/data/sign-language/valid'
test_path = '/home/ggraeff/Documents/Keras/free_code_camp/data/sign-language/test'
```

```python
train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(directory=train_path, class_mode='categorical', target_size=(224,224), batch_size=10)
valid_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(directory=valid_path, class_mode='categorical', target_size=(224,224), batch_size=10)
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(directory=test_path , class_mode='categorical', target_size=(224,224), batch_size=10, shuffle=False)
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
mobile.summary()
```

```python
x = mobile.layers[-5].output
output = Dense(units=10, activation='softmax')(x)
```

```python
model = Model(inputs=mobile.input, outputs=output)
```

```python
for layer in model.layers[:-23]:
    layer.trainable = False
```

```python
model.summary()
```

# compiling and training the model

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x=train_batches, validation_data=valid_batches, epochs=10, verbose=2)
```
